# 1 Himmelblau函数
$f(x,y)=(x^2+y-11)^2+(x+y^2-7)^2$

局部最小值 (3,2) (-2.8,3.1) (-3.76,-3.28) (3.5,-1.85)

```python
def himmelblau(x):
    # x:{x[0],x[1]}
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 1)**2
```

```python
x = np.arange(-6,6,0.1)
y = np.arange(-6,6,0.1)
x.shape,y.shape
```

```python
120*120
```

```python
X,Y=np.meshgrid(x,y)
```

```python
X.shape,Y.shape
```

```python
X,Y
```

```python
Z = himmelblau([X,Y])
```

```python
Z
```

```python
fig = plt.figure('himmelblau')
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z)
ax.view_init(60,-30)
plt.show()
```

```python
#参数初始化
x =tf.constant([4.,0.0])
lr = 0.01
for step in range(201):
    with tf.GradientTape() as tape:
        tape.watch(x)
        y = himmelblau(x)
        grads = tape.gradient(y,x)
    x = x - lr*grads
    if step % 20 == 0:
        print('step {}:x={},f(x):{}'.format(step,x,y))
```
